```python
# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


# Child Class 1: Student
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"{self.name} has enrolled in {course_name}.")

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")


# Child Class 2: Lecturer
class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"{self.name} is teaching {course_name}.")


# Demonstration
if __name__ == "__main__":
    # Create one object of each child class
    student = Student("Angelica", 19, "FOE.41.006.123.25")
    lecturer = Lecturer("Dr. Matthew Cobbinah", 40, "E98765")

    print("Student Information (Overridden display_info)")
    student.display_info()
    student.enroll_course("Object Oriented Programming")

    print("\nLecturer Information (Inherited display_info)")
    lecturer.display_info()
    lecturer.teach_course("Object Oriented Programming")
```
